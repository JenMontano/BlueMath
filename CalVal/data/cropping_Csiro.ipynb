```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

def crop_carolinas_region(ds, lon_min=-80, lon_max=-74, lat_min=31, lat_max=37):
    """
    Crop data to the North and South Carolina region.
    
    Parameters:
    -----------
    ds : xarray.Dataset
        Input dataset with longitude and latitude coordinates
    lon_min : float
        Minimum longitude (default: -80)
    lon_max : float
        Maximum longitude (default: -74)
    lat_min : float
        Minimum latitude (default: 31)
    lat_max : float
        Maximum latitude (default: 37)
        
    Returns:
    --------
    xarray.Dataset
        Cropped dataset for the Carolinas region
    """
    # Convert longitude to 0-360 if needed
    if 'longitude' in ds.coords:
        lon_name = 'longitude'
    elif 'lon' in ds.coords:
        lon_name = 'lon'
    else:
        raise ValueError("No longitude coordinate found in dataset")
        
    if 'latitude' in ds.coords:
        lat_name = 'latitude'
    elif 'lat' in ds.coords:
        lat_name = 'lat'
    else:
        raise ValueError("No latitude coordinate found in dataset")
    
    # Convert negative longitudes to 0-360 if needed
    lon = ds[lon_name].values
    if np.any(lon < 0):
        lon = np.where(lon < 0, lon + 360, lon)
        ds = ds.assign_coords({lon_name: lon})
    
    # Convert input longitudes to 0-360 if needed
    lon_min_360 = lon_min + 360 if lon_min < 0 else lon_min
    lon_max_360 = lon_max + 360 if lon_max < 0 else lon_max
    
    # Crop the dataset
    ds_carolinas = ds.sel(
        {lon_name: slice(lon_min_360, lon_max_360),
         lat_name: slice(lat_min, lat_max)}
    )
    
    return ds_carolinas

def plot_carolinas_region(ds, title="Carolinas Region"):
    """
    Plot the cropped region with coastlines and grid.
    
    Parameters:
    -----------
    ds : xarray.Dataset
        Dataset to plot
    title : str
        Title for the plot
    """
    fig, ax = plt.subplots(figsize=(10, 8),
                          subplot_kw={'projection': ccrs.PlateCarree()})
    
    # Add land and coastlines
    ax.add_feature(cfeature.LAND)
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    
    # Add gridlines
    gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')
    gl.top_labels = False
    gl.right_labels = False
    
    # Plot the data
    if 'longitude' in ds.coords:
        lon_name = 'longitude'
    else:
        lon_name = 'lon'
        
    if 'latitude' in ds.coords:
        lat_name = 'latitude'
    else:
        lat_name = 'lat'
    
    # Plot the first variable in the dataset
    var_name = list(ds.data_vars)[0]
    ds[var_name].plot(ax=ax, transform=ccrs.PlateCarree())
    
    ax.set_title(title)
    plt.show()

# Example usage:
if __name__ == "__main__":
    # Example with a sample dataset
    # Replace this with your actual data loading code
    # ds = xr.open_dataset('your_data.nc')
    
    # Crop to Carolinas region
    # ds_carolinas = crop_carolinas_region(ds)
    
    # Plot the region
    # plot_carolinas_region(ds_carolinas)
    passimport xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

def crop_carolinas_region(ds, lon_min=-80, lon_max=-74, lat_min=31, lat_max=37):
    """
    Crop data to the North and South Carolina region.
    
    Parameters:
    -----------
    ds : xarray.Dataset
        Input dataset with longitude and latitude coordinates
    lon_min : float
        Minimum longitude (default: -80)
    lon_max : float
        Maximum longitude (default: -74)
    lat_min : float
        Minimum latitude (default: 31)
    lat_max : float
        Maximum latitude (default: 37)
        
    Returns:
    --------
    xarray.Dataset
        Cropped dataset for the Carolinas region
    """
    # Convert longitude to 0-360 if needed
    if 'longitude' in ds.coords:
        lon_name = 'longitude'
    elif 'lon' in ds.coords:
        lon_name = 'lon'
    else:
        raise ValueError("No longitude coordinate found in dataset")
        
    if 'latitude' in ds.coords:
        lat_name = 'latitude'
    elif 'lat' in ds.coords:
        lat_name = 'lat'
    else:
        raise ValueError("No latitude coordinate found in dataset")
    
    # Convert negative longitudes to 0-360 if needed
    lon = ds[lon_name].values
    if np.any(lon < 0):
        lon = np.where(lon < 0, lon + 360, lon)
        ds = ds.assign_coords({lon_name: lon})
    
    # Convert input longitudes to 0-360 if needed
    lon_min_360 = lon_min + 360 if lon_min < 0 else lon_min
    lon_max_360 = lon_max + 360 if lon_max < 0 else lon_max
    
    # Crop the dataset
    ds_carolinas = ds.sel(
        {lon_name: slice(lon_min_360, lon_max_360),
         lat_name: slice(lat_min, lat_max)}
    )
    
    return ds_carolinas

def plot_carolinas_region(ds, title="Carolinas Region"):
    """
    Plot the cropped region with coastlines and grid.
    
    Parameters:
    -----------
    ds : xarray.Dataset
        Dataset to plot
    title : str
        Title for the plot
    """
    fig, ax = plt.subplots(figsize=(10, 8),
                          subplot_kw={'projection': ccrs.PlateCarree()})
    
    # Add land and coastlines
    ax.add_feature(cfeature.LAND)
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    
    # Add gridlines
    gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')
    gl.top_labels = False
    gl.right_labels = False
    
    # Plot the data
    if 'longitude' in ds.coords:
        lon_name = 'longitude'
    else:
        lon_name = 'lon'
        
    if 'latitude' in ds.coords:
        lat_name = 'latitude'
    else:
        lat_name = 'lat'
    
    # Plot the first variable in the dataset
    var_name = list(ds.data_vars)[0]
    ds[var_name].plot(ax=ax, transform=ccrs.PlateCarree())
    
    ax.set_title(title)
    plt.show()

# Example usage:
if __name__ == "__main__":
    # Example with a sample dataset
    # Replace this with your actual data loading code
    # ds = xr.open_dataset('your_data.nc')
    
    # Crop to Carolinas region
    # ds_carolinas = crop_carolinas_region(ds)
    
    # Plot the region
    # plot_carolinas_region(ds_carolinas)
    pass
```
