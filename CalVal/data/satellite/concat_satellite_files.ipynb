```python
import os
import xarray as xr
import numpy as np
from time import time

t0 = time()
print('--------------------------------------------------------')
print('Concatenating satellite files and handling duplicates')
print('--------------------------------------------------------')
path = '/Users/sam/jen/code/GeoOcean_codes/CalValWaves/data/satellite'

# First, collect all files
all_files = []
for root, dirs, files in os.walk(
    os.path.join(path, 'sat_netcdf_files_carolinas'), topdown=True
):
    os.chdir(root)
    for name in files:
        if not name.endswith('.nc'):
            continue
        try:
            sat_file = xr.open_dataset(name)
            all_files.append(sat_file)
            if len(all_files) % 10 == 0:
                print(f"Read {len(all_files)} files...")
        except Exception as e:
            print(f'Error reading {name}: {e}')

print(f"\nTotal files read: {len(all_files)}")

# Concatenate and handle duplicates
print("\nConcatenating datasets and removing duplicates...")
merged = xr.concat(all_files, dim='TIME')

# Sort by time
merged = merged.sortby('TIME')

# Remove duplicate times while preserving the data
_, index = np.unique(merged['TIME'], return_index=True)
merged = merged.isel(TIME=index)

# Print coordinate ranges
print("\nCoordinate ranges in merged dataset:")
print(f"Latitude: {merged.LATITUDE.min().values:.2f}° to {merged.LATITUDE.max().values:.2f}°")
print(f"Longitude: {merged.LONGITUDE.min().values:.2f}° to {merged.LONGITUDE.max().values:.2f}°")
print(f"Time: from {merged.TIME.min().values} to {merged.TIME.max().values}")
print(f"Number of time points: {len(merged.TIME)}")

# Check for remaining duplicates
n_times = len(merged.TIME)
n_unique_times = len(np.unique(merged.TIME))
if n_times != n_unique_times:
    print(f"\nWARNING: Still have {n_times - n_unique_times} duplicate timestamps!")
else:
    print("\nNo duplicate timestamps found!")

# Save the merged dataset
output_path = os.path.join('..', 'satellite_dataset_carolinas_test4.nc')
merged.to_netcdf(output_path)
print(f"\nSaved to: {output_path}")
```
