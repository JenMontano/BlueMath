# Generalized Pareto Distribution (GPD)

In this notebook an example of how to use the functions associated to the GPD distribution will be shown

```python
import matplotlib.pyplot as plt

from bluemath_tk.distributions.gpd import GPD
```

Common distribution methods.

```python
threshold = 1.0
scale = 1.0
shape = -0.2
random_state = 42

# Generate a random sample from GEV distribution
x = GPD.random(2000, loc=threshold, scale=scale, shape=shape, random_state=random_state)

x_pdf = GPD.pdf(x, loc=threshold, scale=scale, shape=shape)   # Probability Density Function
x_cdf = GPD.cdf(x, loc=threshold, scale=scale, shape=shape)   # Cumulative Distribution Function
x_ppf = GPD.qf(x_cdf, loc=threshold, scale=scale, shape=shape)  # Percent-Point Function (Quantile Function)
x_sf = GPD.sf(x, loc=threshold, scale=scale, shape=shape)     # Survival Function
x_stats = GPD.stats(loc=threshold, scale=scale, shape=shape)  # Mean, Variance, Skewness, Kurtosis
x_stats
```

Fitting the data

```python
fit_result = GPD.fit(data = x - threshold, f0 = 0.0)  # Fit the GEV distribution to the data
fit_result.summary()  
```

```python
fig = fit_result.plot()  # Plot the fitted GEV distribution against the data
plt.show(fig)  # Show the plot
```
